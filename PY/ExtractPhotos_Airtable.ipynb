```python

import pandas as pd
import numpy as np

path = r"D:\WORK\FLOODNET\WatchData\FloodWatch_Airtable\Airtable_Survey_wPhotoLinks.csv"

df = pd.read_csv(path)

df.head()
```

```python
import requests
import os
import time

image_path = r'D:\WORK\FLOODNET\WatchData\FloodWatch_Airtable\Airtable_Photos'

def download_image(image_url, image_name):
    img_data = requests.get(image_url).content
    with open(image_name, 'wb') as handler:
        handler.write(img_data)


for i,r in df.iterrows():

    if i > 483:
        photo_links = r['Photo']

        if not pd.isna(r['Photo']):

            photo_links = [i for i in r['Photo'].split( "," ) if "airtableuser" in i]
            unique_id = r['unique_id']

            for count,link in enumerate(photo_links):
                single_link = link.replace( '(1)' , '' )
                single_link = single_link.split( "(")[1].split( ")")[0]
                download_image( single_link , os.path.join( image_path , f"{unique_id}-{count}.jpg") )

        time.sleep(0.1)

print( "done" )
```
