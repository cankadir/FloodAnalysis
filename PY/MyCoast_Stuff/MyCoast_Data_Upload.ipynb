```python
import pandas as pd
import os

def isnan(value):
  try:
      import math
      return math.isnan(float(value))
  except:
      return False


survey_link = r"D:\WORK\FLOODNET\WatchData\FloodWatch_Airtable\Airtable_Survey_wPhotoLinks.csv"
photos_folder = r"D:\WORK\FLOODNET\WatchData\FloodWatch_Airtable\Airtable_Photos_Convert"

df = pd.read_csv(survey_link)
df = df.drop(['Photo','S123_Object_ID'], axis=1 )
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python

df['What is affected by the flood waters?'] = df['What is affected by the flood waters?'].fillna('')

a = []
for i,r in df['What is affected by the flood waters?'].items():
    if r == '':
        a.append('')
    elif isnan(r) == True:
        a.append('')
    else:
        all_enteries = r.split(',')

        temp_list = []
        for entry in all_enteries:
            temp = entry.strip()

            if temp in [ 'Roads streets','roads_streets','Roadway/street', 'Roadway/ street', 'Roads/streets' ]: temp = 'Roads/streets'
            elif temp in ['sidewalk', 'sidewalks', 'Sidewalk', 'Sidewalks' ] : temp = 'Sidewalks'
            elif temp in ['parking lot', 'parking lots', 'Parking lotss',"Parking_Lots","parking_lots","Parking_Lots",'Parking lots']: temp = 'Parking lots'
            elif temp in ['property',"Structures"]: temp = "Structures"
            elif temp in ['other',"Other"]: temp = "Other"
            else: temp = temp
            
            temp_list.append(temp)

        a.append( temp_list )

df['What is affected by the flood waters?'] = a

# Other Separate
allowed_options = [ 'Roads/streets', 'Sidewalks' , 'Parking lots', 'Lawns/vegetation', 'Structures', 'Other']

allowed_text = []
not_allowed_text = []

for i,r in df['What is affected by the flood waters?'].items():
    if r == '':
        allowed_text.append('')
        not_allowed_text.append('')
    else:
        temp_allowed = []
        temp_not_allowed = []

        for entry in r:
            if entry in allowed_options:
                temp_allowed.append(entry)
            else:
                temp_not_allowed.append(entry)
        
        allowed_text.append( temp_allowed )
        not_allowed_text.append( temp_not_allowed )

df['What is affected by the flood waters?'] = allowed_text
df['What is affected by the flood waters?_other'] = not_allowed_text

df['What is affected by the flood waters?'] = ['' if len(x)==0 else x for x in df['What is affected by the flood waters?'].values]

df['What is affected by the flood waters?_other'] = df['What is affected by the flood waters?_other'].apply( lambda x: ",".join(x) if len(x)>0 else "" ) 

t = df[ df['What is affected by the flood waters?'] != '' ]['What is affected by the flood waters?']
t = t.explode()

t.unique()
```

# Resize images  
Make sure **NOT** to work on the original images, create a new resize_folder.  
This code overrides the exsisting files in the folder. 

`max_size` defines the max size allowed on either dimension of the image.

```python
from PIL import Image
max_size = 720

# There are reports without images attached to them.
for i,r in df[df['Photo_filenames']!='' ].iterrows():

    row_id = r['unique_id']
    row_photos = r['Photo_filenames'].split(', ')

    row_photos_fullpath = [ os.path.join(photos_folder, i) for i in row_photos ]

    for img_path in row_photos_fullpath: 
        if img_path != '': 

            image = Image.open( img_path )

            # Image dimensions 
            x = image.size[0]
            y = image.size[1]
            ratio = x/y
            print( img_path )

            if (x > max_size) | (y > max_size ):
                # Make sure the image is RGB. 
                image = image.convert('RGB')

                # limit longest side to 1000px
                if x > y:
                    new_x = max_size
                    new_y = int(max_size/ratio)
                else:
                    new_y = max_size
                    new_x = int(max_size*ratio)

                try:
                    image_resized = image.resize( (new_x, new_y) )
                    image_resized.save( img_path )
                except:
                    print( "Error resizing image: ", img_path )
                    image.save( img_path )
                    
```
